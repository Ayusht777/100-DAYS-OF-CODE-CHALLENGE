## Operator Overloading

This feature in Python that allows the same operator to have different meaning according to the context is called operator overloading.

#### Problem 1
Fill in the Line class methods to accept coordinates as a pair of tuples and return the slope and distance of the line.

```python
class Line():
    # the input will be tuple
    def __init__(self,cordinate1,cordinate2):
         self.cordinate1 = cordinate1
         self.cordinate2 = cordinate2
    def distance(self):
        x1,y1 =self.cordinate1
        x2,y2 = self.cordinate2
        return(((x2-x1)**2 + (y2-y1)**2)**0.5)
    def slope(self):
        x1,y1 =self.cordinate1
        x2,y2 = self.cordinate2
        return (y2-y1)/(x2-x1)
        
```

```python
c1 = (0,5)
c2 = (1,8)
line =Line(c1,c2)
```

```python
line.distance()
```

```python
line.slope()
```

```python
class Cylinder:
    
    def __init__(self,height=1,radius=1):
        pass
        
    def volume(self):
        pass
    
    def surface_area(self):
        pass
```

```python
class Cylinder:
    PI = 3.14
    
    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius =radius
        
    def volume(self):
        h = self.height
        r = self.radius
        return(self.PI*h*r*r)   
    def surface_area(self):
        h = self.height
        r = self.radius
        return(2*self.PI*r*h +2*self.PI*r*r) 
        
```

```python
c = Cylinder(5,5)
```

```python
c.volume()
```

```python
c.surface_area()
```

# Object Oriented Programming Challenge

For this challenge, create a bank account class that has two attributes:

* owner
* balance

and two methods:

* deposit
* withdraw

As an added requirement, withdrawals may not exceed the available balance.

Instantiate your class, make several deposits and withdrawals, and test to make sure the account can't be overdrawn.

```python
class Bank():
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance =balance
        
    def __str__(self):
        return(f"Owner Name = {self.owner}\nBalance = {self.balance}")
    def deposit(self,amount):
        self.balance = amount + self.balance
        print(f"FUNDS ACCEPTED")
    def withdraw(self,amount_w):
        if(amount_w<=self.balance):
            self.balance = self.balance - amount_w
            print(f"WITHDRAW AMOUNT : {amount_w}")
        else:
            print('YOUR BALANCE IS LOW')
```

```python
b = Bank('ayush',500)
```

```python
b.withdraw(100)
```
