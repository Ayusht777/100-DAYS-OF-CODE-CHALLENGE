## Encapsulation
When working with classes and dealing with sensitive data, providing global access to all the variables used within the program is not a good choice. Encapsulation offers a way for us to access the required variables without providing the program full-fledged access to any of those variables.

Updating, modifying, or deleting data from variables can be done through the use of methods that are defined specifically for the purpose. The benefit of using this approach to programming is improved control over the input data and better security.

in another word protecting the class data

```python
# data without Encapsulation
class person:
    def __init__(self,name,age,salary):
        self.name =name
        self.age = age
        self.salary = salary
    def person_name(self):
        print(self.name)      
```

```python
p1 =person(name='ayush',age=25,salary=120000000)
```

```python
p1.person_name()# this will print the name of the person
```

```python
p1.age
```

```python
p1.salary# so you can easily access the data so to stop it we use Encapsulation
```

### 1. Method for Encapsulation
##### Using Single Underscore
use _ (single underscore) before the varible name


```python
class person:
    def __init__(self,name,age,salary):
        self.name =name
        self._age = age# it is still callable but it will tell another developer that it is  aprivate varibale
        self._salary = salary
    def person_name(self):
        print(self.name)    
```

```python
p2 = person('ayush',24,120000000)
```

```python
p2.person_name()
```

```python
p2._age
```

```python
p2._salary
```

### 2. Method for Encapsulation
##### Using Double Underscore
use __ (Double underscore) before the varible name


```python
class person:
    def __init__(self,name,age,salary):
        self.name =name
        self.__age = age
        self.__salary = salary
    def person_name(self):
        print(self.name)    
```

```python
p3 = person('ayush',26,175200000)
```

```python
p3.person_name()
```

```python
p3.__age# now it is private
```

```python
p3.__salary
```

### Name Mangling Method
We can directly access private and protected variables from outside of a class through name mangling. The name mangling is created on an identifier by adding `__` two leading underscores and one trailing underscore, like this `_classname__dataMember`, where `classname` is the current class, and data member is the private variable name.

`syntax : assignedobj._classname__variable `

```python
p3._person__salary
```

```python
p3._person__age
```

```python
p3._person__age =55 # we can also modifiy it
```

```python
p3._person__age
```

### Getters and Setters Method

If you want to access and change the private variables, accessor (getter) methods and mutators(setter methods) should be used, as they are part of Class.

```python
class person:
    def __init__(self,name,age=0,salary=0):
        self.name =name
        self.__age = age
        self.__salary = salary
    def person_name(self):
        print(self.name)  
        
    def get_salary(self):
        print(self.__salary)  
    def set_salary(self,salary):
        self.__salary =salary
```

```python
p4 =person('Dev', 30,555666666)
```

```python
p4.person_name()
```

```python
p4.get_salary()
```

```python
p4.set_salary(644646464646)
```

```python
p4.get_salary()
```

### Polymorphism 

The word polymorphism means having many forms. In programming, polymorphism means the same function name (but different signatures) being used for different types.

```python
Image("https://cdn.programiz.com/sites/tutorial2program/files/func-polymorphism.png",width=400,height=400)
```

```python
print(len("ayush"))
print(len([1,2,3,4]))
# same function name but different use scenario
```

```python
class India:
    def capital(self):
        print('The Capital of india is New Delhi')
    def language(self):
        print("Hindi is the most widely spoken language of India.")
    def currency(self):
        print('The currency os india is rupee')
class Japan:
    def capital(self):
        print('The Capital of Japan is Tokoyo')
    def language(self):
        print("japanense is the most widely spoken language of Japan.")
    def currency(self):
        print('The currency os japna is Yen')
```

```python
i = India()
```

```python
u =Japan()
```

```python
print(i.capital())
print(u.capital())
```

```python
i.capital()
u.capital()
```

```python
for x  in i,u:
    x.capital()
    x.currency()
    x.language()
    print("-----------------------------------")
```

```python
for x  in i,u:
    print(type(x))
    x.capital()
    x.currency()
    x.language()
   
    print("-----------------------------------")
```

##### Polymorphism with Inheritance

```python
class Animal:
    def type(self):
        print('i am a animal')
class snake(Animal):
    def type(self):
        print('i am a snake')
class python(snake):
    
    def type(self):
        print("i am a python ")
```

```python
a= Animal()
s =snake()
p = python()
```

```python
for x in a,s,p:
    x.type()
    print(type(x))
```
