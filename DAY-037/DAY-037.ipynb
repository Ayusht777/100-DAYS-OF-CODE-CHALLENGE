## $ – Dollar

or 

\Z

> Dollar($) symbol matches the end of the string i.e checks whether the string ends with the given character(s) or not. For example – 

    s$ will check for the string that ends with a such as geeks, ends, s, etc.
    ks$ will check for the string that ends with ks such as geeks, geeksforgeeks, ks, etc.

```python
import re
x  = 'ayush talesara'
```

```python
re.search('ra$',x)
```

```python
re.search('ar$',x)
```

```python
if (re.search("[0-9]",x)):
    print("f")
else:
    print("nf")
```

```python
if (re.search("ra\Z",x)):
    print("f")
else:
    print("nf")
```

# . – Dot

> Dot(.) symbol matches only a single character except for the newline character (\n). For example –  

    a.b will check for the string that contains any character at the place of the dot such as acb, acbd, abbb, etc
    .. will check if the string contains at least 2 characters

```python
x = 'ayush t'
```

```python
re.search('ayush.t',x)
```

```python
print(re.search('ayush.t',"ayush\nt"))
```

# | – Or

> Or symbol works as the or operator meaning it checks whether the pattern before or after the or symbol is present in the string or not. For example –  

    a|b will match any string that contains a or b such as acd, bcd, abcd, etc.

```python
re.search('a|t',x)
```

```python
re.search('t|z',x)
```

# ? – Question Mark

> Question mark(?) checks if the string before the question mark in the regex occurs at least once or not at all. For example –  

    ab?c will be matched for the string ac, acb, dabc but will not be matched for abbc because there are two b. Similarly, it will not be matched for abdc because b is not followed by c.

```python
re.search("ayush?",x)
```

```python
print(re.search("q-?",x))
```

# * – Star

>Star (*) symbol matches zero or more occurrences of the regex preceding the * symbol. For example –  

    ab*c will be matched for the string ac, abc, abbbc, dabc, etc. but will not be matched for abdc because b is not followed by c.

```python
re.search("ab*c","abcde")
```

```python
re.search("ba*c","abcde")
```

```python
re.search("c*a","abcde")
```

# + – Plus

> Plus (+) symbol matches one or more occurrences of the regex preceding the + symbol. For example –  

    ab+c will be matched for the string abc, abbc, dabc, but will not be matched for ac, abdc because there is no b in ac and b is not followed by c in abdc.

```python
print(re.search('foox+bar', 'fooxxbar'))              # Zero dashes


print(re.search('foo0+bar', 'foo-bar') )                  # One dash


print(re.search('=+bar', 'foo--bar'))  
```

# {m}

>    Matches exactly m repetitions of the preceding regex.

This is similar to * or +, but it specifies exactly how many times the preceding regex must occur for a match to succeed:

```python
print(re.search("-{4}",'a----x'))
print(re.search("-{3}",'a----x'))
print(re.search("a{0}",'a----x'))
```

# {m, n} – Braces

> Braces match any repetitions preceding regex from m to n both inclusive. For example –  

    a{2, 4} will be matched for the string aaab, baaaac, gaad, but will not be matched for strings like abc, bc because there is only one a or no a in both the cases.

```python
for i in range(1,7):
    s = f"@#{'a' * i}@"
    print(f'{i}  {s:15}', re.search('@#a{2,7}', s))
```

# (< regex >) – Group

> Group symbol is used to group sub-patterns. For example –  

    (a|b)cd will match for strings like acd, abcd, gacd, etc.

```python
print(re.search("(ush)",x))
print(re.search("ush",x))
```

# \A similar to anchor

# \b	
> Matches if the word begins or ends with the given character. \b(string) will check for the beginning of the word and (string)\b will check for the ending of the word.

```python
print(x)
print(re.search(r"ayush\b",x))
print(re.search("ayush\b",x))
print(re.search(r"ayush\b t",x))
```

# \B	
> It is the opposite of the \b i.e. the string should not start or end with the given regex.

```python
print(x)
print(re.search(r"s\B",x))
print(re.search("y\B",x))
print(re.search(r"a\B t",x))
print(re.search(r"ayush\B",x))
print(re.search("ayush\B",x))
print(re.search(r"ayush\B t",x))
```

# \d	
> Matches any decimal digit, this is equivalent to the set class [0-9]

```python
print(re.search("\d",'546464a'))
print(re.search("\d",x))
print(re.search(r"\d",'as464dsadasd'))
```

# \D	
>Matches any non-digit character, this is equivalent to the set class [^0-9]

```python
print(re.search("\D",'546464'))
print(re.search("\D",x))
print(re.search(r"\D",'as464dsadasd'))
```

# \s	
>Matches any whitespace character.

```python
print(re.search("\s",'ayush t'))
print(re.search(r"\s","ayush t"))
print(re.search("\s",'agsga'))
```

# \S	
>Matches any non-whitespace character

```python
print(re.search("\S",'546464'))
print(re.search(r"\S",x))
print(re.search(r"\S",'\n'))
print(re.search("\S",'as46!4ddasd'))
```

# \w	
>Matches any alphanumeric character, this is equivalent to the class [a-zA-Z0-9_].

```python
print(re.search('\w','!@##$%&*()'))
print(re.search('\w','!@##$%&*(a)'))
print(re.search(r'\w','!@##$%&*(  )'))
```

# \W	
>Matches any non-alphanumeric character.

```python
print(re.search('\W','!@##$%&*()'))
print(re.search('\W','aysuh'))
print(re.search(r'\W','!@##$%&*(  )'))
```

# re.Match( )
## re.match(< regex >, < string >, flags=0)
>match.re attribute returns the regular expression passed and match.string attribute returns the string passed.


>This is identical to re.search(), except that re.search() returns a match if <regex> matches anywhere in <string>, whereas re.match() returns a match only if <regex> matches at the beginning of <string>:


```python
print(re.match('1','12ad'))
print(re.match('2','12ad'))
print(re.match(r'1','12ad'))
print(re.match(r'2','12ad'))
print(re.match('\d','12ad'))
```

# re.fullmatch(< regex >, < string >, flags=0)

    Looks for a regex match on an entire string.

> This is similar to re.search() and re.match(), but re.fullmatch() returns a match only if < regex > matches < string > in its entirety

### < findstring > = < string >

```python
print(re.fullmatch("123",'123'))
print(re.fullmatch(r"123",'123'))
print(re.fullmatch("321",'123a'))
print(re.fullmatch("\d",'123a'))
```

# re.findall(< regex >, < string >, flags=0)

    Returns a list of all matches of a regex in a string.

> re.findall(< regex >, < string >) returns a list of all non-overlapping matches of < regex > in < string >. It scans the search string from left to > > right and returns all matches in the order found

```python
print(re.findall("[a-z]","!@@#$%^&*Ayush&*Talesara,asa"))
print(re.findall("[A-Z]","!@@#$%^&*Ayush&*Talesara,asa"))
print(re.findall("[0-9]","!@@#$%^&*Ayush&*Talesara,asa"))
print(re.findall(r"[a-z]","!@@#$%^&*Ayush&*Talesara,asa"))
print(re.findall(r"\w","!@@#$%^&*Ayush&*Talesara,asa"))
print(re.findall("\w+","!@@#$%^&*Ayush&*Talesara,asa"))
print(re.findall("(\W+)","!@@#$%^&*Ayush&*Talesara,asa"))
print(re.findall("(\w+),(\w+)","!@@#$%^&*Ayush&*Talesara,asa"))
print(re.findall(r"(\W+),(\W+)","!@@#$%^&*Ayush&*Talesara,asa"))
```

>re.finditer(< regex >, < string >) scans < string > for non-overlapping matches of < regex > and returns an iterator that yields the match objects from any it finds. It scans the search string from left to right and returns matches in the order it finds them:
