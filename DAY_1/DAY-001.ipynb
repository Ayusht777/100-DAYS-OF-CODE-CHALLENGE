```python
#strings
a = "abcd"
b = "efgh"
print(a+b)
#print(a-b) we cannot sub the two string
c = a+b #Concatenation of Two or More Strings
print(c)
print(type(c))
print(len(c))

print ("c = ",c)

# string indexing and slicing 
print(c)
print(c[0])
print(c[5])
print(c[-2])
print(c[0:2])
print(c[0:8:2]) # syntax [start:stop:step]
print(c[::-1]) # reverse string
#print(c[1.4],c[17]) the indexing number should be a int and <=len(string)

#string properties
x = "ayush"
y = "talesara"
z = x + y[0]
print(z)

# string manuplation
# formating with %
# .format() method
# f-string method

#---------------------------------------------------------------------------------------
# % formating 
name = "ayush"
print("what is your name : %s"%(name))
# it just like c 
number = 475
print("the number = %d,%s"%(number,number))
# both the %s and %d will work
f = 1.454
print("f = %f"%(f))
# percsion 
print("f = %.2f"%(f))
# padding
print("f =%40.2f"%(f))
# %r 
print("raw = %r"%(name)) # raw method

# the % formating is old method 


#.format() method 

n = "ayush talesra"
num = 7894894894778
print("your name is {}".format(n))# it is good because you donot need to define data type in place holder
print("your name is {} and number = {}".format(n,num))
# just repeated any var
print("{0} {0}".format(n))
print("{1} {0}".format(n,num)) # oder

#float formatting
# syntax {value:spcae.precision}
z = 78.5855184
print("number = {:10.2f}".format(z))
```

```python
#Alignment, padding and precision with `.format()`
print ("name = {:>10}".format(name)) # left spacing
print ("name = {:<10} :".format(name)) # right spacing
print ("name = {:^10}".format(name)) # center spacing




```
